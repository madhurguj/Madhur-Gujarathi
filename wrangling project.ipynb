```python
df=df[["continent","location","date","total_cases","total_deaths","gdp_per_capita","human_development_index"]]
df
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python

df.isnull().sum()
```

```python
df=df.dropna(subset=["continent"])
```

```python
df.isnull().sum()
```

```python
df=df.fillna(0)
```

```python
df.isnull().sum()
```

```python
df["date"]=pd.to_datetime(df["date"])
df.head(5)
```

```python
df["month"]=pd.to_datetime(df["date"]).dt.month
df
```

```python
df_max=df.groupby(["continent"]).max()
df_max
```

```python
df_groupby=df_max.reset_index()
df_groupby
```

```python
df_groupby["total_deaths_to_total_cases"]=df_groupby["total_deaths"]/df_groupby["total_cases"]
df_groupby
```

```python
sns.distplot(df_groupby["gdp_per_capita"],hist=True, kde=False)
```

```python
 sns.jointplot(data= df_groupby, x="total_cases", y="gdp_per_capita", kind='scatter')
```

```python
sns.pairplot(data=df_groupby, hue="continent")
```

```python
sns.barplot(data=df_groupby, x='continent', y='total_cases')
```
